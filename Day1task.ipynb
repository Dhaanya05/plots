```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,11)
y=x
print(x)
print(y)
plt.figure(figsize=(6,4))
plt.plot(x,y,marker="X",linestyle="--",color="pink")
plt.title("Line plot")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.grid(True)
plt.show()
```

```python
x=['A','B','C','D']
y=[40,70,30,90]
plt.figure(figsize=(6,4))
plt.bar(x,y,color="skyblue")
plt.title("Bar plot")
plt.xlabel("Products")
plt.ylabel("values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(11)

x = np.arange(1, 11)
y = np.random.randint(15, 100, size=10)

plt.figure(figsize=(10, 12))

plt.subplot(2, 2, 1)
plt.plot(x, y, marker="o", color="blue")
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)


plt.subplot(2, 2, 2)
plt.scatter(x, y, color="red", marker="s")
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Random Values")
plt.grid(True)


plt.subplot(2, 2, 3)
plt.bar(x, y, color="green")
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")


plt.subplot(2, 2, 4)
plt.hist(y, bins=10, color="pink")
plt.title("Histogram")
plt.xlabel("Time")
plt.ylabel("Frequency")

plt.show()
```
