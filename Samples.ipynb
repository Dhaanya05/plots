```python
import numpy as np
import pandas as pd
np.random.seed(11)
data={
    "x" : np.arange(1,10),
    "y" : np.random.randint(15,100,size=9)
}
df=pd.DataFrame(data)
print(df)

import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
x=df["x"]
y=df["y"]
plt.plot(x,y,marker="*",color="blue")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Random plot")
plt.grid(True)
plt.show()
```

```python
plt.ylabel("Categories")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(x,y,color='pink')
plt.title("Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```
